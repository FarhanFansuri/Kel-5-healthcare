# Visualisasi

```python
# Dapatkan nilai counts dari efek samping dan batasi hingga 10 teratas
efek_samping_counts = clean_df['Side_effects'].value_counts().head(10)

# Buat kamus untuk menyimpan obat-obatan untuk setiap efek samping
efek_samping_obat = {}

# Loop melalui efek samping unik dan kumpulkan obat yang terkait
for efek_samping in efek_samping_counts.index:
    # Filter DataFrame untuk setiap efek samping unik
    obat_untuk_efek_samping = clean_df.loc[clean_df['Side_effects'] == efek_samping, 'Medicine Name'].tolist()

    # Simpan daftar obat untuk efek samping dalam kamus
    efek_samping_obat[efek_samping] = obat_untuk_efek_samping

# Buat grafik batang untuk 10 efek samping teratas dan jumlah kemunculannya dengan estetika yang ditingkatkan
plt.figure(figsize=(12, 8))
grafik_batang = efek_samping_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Menambahkan anotasi untuk setiap bar dengan format kustom
for i, jumlah in enumerate(efek_samping_counts):
    plt.text(i, jumlah + 1, f'{jumlah}', ha='center', va='bottom', fontsize=10)

# Gambar garis horizontal untuk jumlah rata-rata
jumlah_rata_rata = efek_samping_counts.mean()
plt.axhline(jumlah_rata_rata, color='red', linestyle='--', label=f'Rata-rata: {jumlah_rata_rata:.2f}')
plt.legend()

# Gambar garis horizontal untuk jumlah median
jumlah_median = efek_samping_counts.median()
plt.axhline(jumlah_median, color='green', linestyle="--", label=f'Median: {jumlah_median:.2f}')
plt.legend()

plt.title('Top 10 Kemunculan Efek Samping')
plt.xlabel('Efek Samping')
plt.ylabel('Jumlah')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Dapatkan nilai counts dari penggunaan (Uses)
penggunaan_counts = clean_df['Uses'].value_counts().head(10)

# Buat kamus untuk menyimpan obat-obatan untuk setiap penggunaan
penggunaan_obat = {}

# Loop melalui penggunaan unik dan kumpulkan obat yang terkait
for penggunaan, hitungan in penggunaan_counts.items():
    # Filter DataFrame untuk setiap penggunaan unik
    obat_untuk_penggunaan = clean_df.loc[clean_df['Uses'] == penggunaan, 'Medicine Name'].tolist()

    # Simpan daftar obat untuk penggunaan dalam kamus
    penggunaan_obat[penggunaan] = obat_untuk_penggunaan

# Tampilkan hanya 10 penggunaan pertama dan obat-obat yang terkait
for idx, (penggunaan, obat) in enumerate(penggunaan_obat.items()):
    if idx >= 10:
        break  # Keluar dari loop setelah mencetak 10 kemunculan

    print(f"Penggunaan: {penggunaan}")
    print(f"Jumlah Kemunculan: {penggunaan_counts[penggunaan]}")
    print(f"Obat-obatan: {obat}")
    print("------------------------")
```

```python
# Ambil 10 penggunaan teratas
top_10_penggunaan = penggunaan_counts.head(10)

# Hitung nilai median dan mean
nilai_median = top_10_penggunaan.median()
nilai_rata_rata = top_10_penggunaan.mean()

# Buat grafik batang untuk 10 penggunaan teratas dan jumlah kemunculannya
plt.figure(figsize=(13, 10))
grafik_batang = top_10_penggunaan.plot(kind='bar', color='skyblue', edgecolor='black')

# Menambahkan anotasi untuk setiap bar
for i, jumlah in enumerate(top_10_penggunaan):
    plt.text(i, jumlah + 1, str(jumlah), ha='center', va='bottom', fontsize=10)

# Plot garis median dengan anotasi
plt.axhline(nilai_median, color='red', linestyle='--', linewidth=1.5, label=f'Median: {nilai_median:.2f}')
plt.text(len(top_10_penggunaan) - 1, nilai_median, f'Median: {nilai_median:.2f}', color='red', va='bottom', ha='right', fontsize=10)

# Plot garis mean dengan anotasi
plt.axhline(nilai_rata_rata, color='green', linestyle='-.', linewidth=1.5, label=f'Rata-rata: {nilai_rata_rata:.2f}')
plt.text(len(top_10_penggunaan) - 1, nilai_rata_rata, f'Rata-rata: {nilai_rata_rata:.2f}', color='green', va='bottom', ha='right', fontsize=10)

plt.title('Top 10 Kemunculan Penggunaan dan Obat-obatan Terkait', fontsize=14)
plt.xlabel('Penggunaan', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, top_10_penggunaan.max() * 1.1)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
