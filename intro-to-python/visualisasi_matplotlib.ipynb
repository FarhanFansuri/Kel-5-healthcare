## 2. Contoh penggunaanya

### import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### dataset gathering

```python
tahun = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026])
penjualan = np.array([100, 120, 150, 200, 250, 300, 320, 300, 280, 250, 200, 180])
```

### plotting sederhana

```python
plt.plot(tahun,penjualan)
plt.title("Jumlah penjualan")
plt.xlabel("tahun")
plt.ylabel("penjualan")
# plt.xticks(tahun)
# plt.yticks(penjualan)
plt.show()
```
